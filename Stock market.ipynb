```python
import yfinance as yf
import pandas as pd

stocks = [
    "AAPL",
    "MSFT",
    "NVDA",
    "AMZN",
    "GOOGL",
    "META",
    "TSLA",
    "NFLX",
    "JPM",
    "WMT"
]

all_data = []

for stock in stocks:

    print(f"Downloading {stock}")

    df = yf.download(
        stock,
        start="2020-01-01",
        end="2026-06-30",
        progress=False,
        auto_adjust=False
    )

    df = df.reset_index()

    # Flatten MultiIndex columns if present
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df["Company"] = stock

    all_data.append(df)

stock_data = pd.concat(all_data, ignore_index=True)

stock_data = stock_data[
    [
        "Date",
        "Company",
        "Open",
        "High",
        "Low",
        "Close",
        "Volume"
    ]
]

stock_data.to_csv("stock_market_cleaned.csv", index=False)

print(stock_data.head())
```

```python
numeric_columns = stock_data.select_dtypes(include="number").columns

stock_data[numeric_columns] = stock_data[numeric_columns].fillna(
    stock_data[numeric_columns].median()
)

stock_data["Company"] = stock_data["Company"].fillna("Unknown")
```

```python
stock_data.drop_duplicates(inplace=True)
```
